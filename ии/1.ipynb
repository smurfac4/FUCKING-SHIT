```python
import math as m

x = float(input("x= "))
y = float(input("y = "))
num = int(input("Введите номер операции \n1) +\n2) -\n3) \n4) /\n5) e^x+y\n6) sin(x+y)\n7) cos(x+y)\n8) x^y\n"))

if num == 1:
    print (x ," + ", y , " = " , x + y)
elif num == 2:
    print (x ," - ", y , " = ", x - y)
elif num == 3:
    print (x ," ", y , " = ", x * y)
elif num == 4:
    print (x ," / ", y , " = ", x / y)
elif num == 5:
    print ("e ^ (", x ," + ", y , ") = ", m.exp(x+y))
elif num == 6:
    print ("sin(", x, " + " , y, ") =", m.sin(x+y))
elif num == 7:
    print ("cos(", x, " + " , y, ") =", m.cos(x+y))
elif num == 8:
    print (x,"^", y ,"=" , x**y)
else:
    print("Введено неккоректный символ")
```
